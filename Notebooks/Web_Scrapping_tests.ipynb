Over here we will test ways to Web-Scrap yfinance website

```python
import requests
from bs4 import BeautifulSoup

def fetch_trending_yahoo():
    url = "https://finance.yahoo.com/trending-tickers"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
    }
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    trending_tickers = []

    # find the trend table rows
    rows = soup.select("table tr")

    for row in rows[1:]:  # skip header row
        cols = row.find_all("td")
        if len(cols) >= 2:
            ticker = cols[0].text.strip()
            name = cols[1].text.strip()
            trending_tickers.append({"ticker": ticker, "name": name})

    return trending_tickers

# test
print(fetch_trending_yahoo())
```
